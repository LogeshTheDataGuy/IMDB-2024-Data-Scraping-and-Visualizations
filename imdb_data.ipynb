```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import time

for genre in ['action','adventure','crime','documentary','horror','romance']:

    driver=webdriver.Chrome()
    url=f'https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31&genres={genre}'
    driver.get(url)

    time.sleep(3)

    more_button = driver.find_element(By.XPATH,'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/div[2]/div/span')

    def click_more_button():
        try:
            action=ActionChains(driver)
            action.move_to_element(more_button).perform()
            more_button.click()
            time.sleep(2)
            return True
        
        except Exception as e:
           # print(f"no more next button:{e}")
            return False

    while click_more_button():
        print("50 more button available")

    print("50 more button not available")

    movies=[]
    durations=[]
    ratings=[]
    votings=[]

    time.sleep(3)

    movie_items=driver.find_elements(By.XPATH,'//*[@id="__next"]/main/div[2]/div[3]/section/section/div/section/section/div[2]/div/section/div[2]/div[2]/ul/li')

    for movie_item in movie_items:
        try:
            movie=movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/div[1]/a/h3').text
            duration=movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/div[2]/span[2]').text
            rating=movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/span/div/span/span[1]').text
            voting=movie_item.find_element(By.XPATH,'./div/div/div/div[1]/div[2]/span/div/span/span[2]').text

            movies.append(movie)
            durations.append(duration)
            ratings.append(rating)
            votings.append(voting)

        except Exception as e:
           #print(f'Error extracting the data :{e}')
            continue

    df=pd.DataFrame(
    {
        'Movie':movies,
        'Duration':durations,
        'Rating':ratings,
        'Votes':votings
        }
        )

    driver.quit()
    df.to_csv(f'{genre}1.csv')
    df
```

```python

def convert_to_minutes(duration):
    parts = duration.split() 
    hours = 0
    minutes = 0

    for part in parts:
        if 'h' in part: 
            hours = int(part.replace('h', ''))  
        elif 'm' in part: 
            minutes = int(part.replace('m', ''))  
    
    return (hours * 60) + minutes  

df['Duration'] = df['Duration'].apply(convert_to_minutes)
df.rename(columns={'Duration':'Duration(Mins)'},inplace=True)
df
```
